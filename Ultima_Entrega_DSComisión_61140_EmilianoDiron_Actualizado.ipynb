```python
# Sabemos que las primeras 2 columnas no aportan al dataset, procedemos a eliminarlas
df = df.drop(['Unnamed: 0', 'id'], axis=1)
# Comprobamos el resultado
df.info()
```

# Estrategias para manejar valores en 0 y vacios

```python
# Por descripcion de columnas sabemos cuales permiten valores en 0, lo cual significa que no fue evaluado por el pasajero
# Para estos casos definimos pasar los valores en 0 a nam para luego manejar los nam.
# Primero creamos una lista con todos los nombres de columna donde podemos realizar esta accion
satisfaction_cols = [
    'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking',
    'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment',
    'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service',
    'Inflight service', 'Cleanliness'
]
# Luego reemplazamos los valores 0 por NaN
df[satisfaction_cols] = df[satisfaction_cols].replace(0, np.nan)

#Luego de realizar el cambio, procedemos a verificar cuantos nam hay por columna para determinar la estrategia a seguir.
df.isna().sum()
```

```python
# Definimos como estrategias específicas para manejar NaN reemplazarlos por la media de todas las columnas ya que en ningun caso supera el 5% de la cantidad de datos.

#Guardamos la media de cada columna como INT
mean_values = df[satisfaction_cols].mean().round().astype(int)

#Reemplazmos los valores con las medias calculadas
for col in satisfaction_cols:
    df[col].fillna(mean_values[col], inplace=True)

# Verificamos el cambio
df.isna().sum()
```

```python
# Los unicos datos que nos quedan en NAM son de la columna Arrival Delay in minutes
# En este caso vamos a eliminar los registros donde la columna Arrival delay in minutes es nam

df.dropna(inplace=True)

# Verificamos el cambio
df.isna().sum()
```

## Seleccion del de las Feautere seleccion y Algoritmo

Para poder definir las variables importantes, primero elegi el Algoritmo de RandomForest, el cual consiste de arboles logicos y una de las ventajas que nos proporciona es la medida de imporatancia de cada variable, con ello podemos elegir nuestras variables importantes.

```python
# Librerias necesarias

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Convertir variables categóricas a numéricas

label_encoders = {}
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
# Separo las variables que van a ayudar a predecir la variable objetivo (sexo, edad etc) de la variable objetivo (satisfaccion)
X = df.drop(columns=['satisfaction'])
y = df['satisfaction']
```

```python
# Divido los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
# Entreno un modelo seleccionado
rf_model = RandomForestClassifier(random_state=30)
rf_model.fit(X_train, y_train)
```

```python
# Cuales variables son mas importantes (cuales tienen mas peso)
feature_importances = pd.DataFrame(rf_model.feature_importances_, index=X_train.columns, columns=['importance']).sort_values('importance', ascending=False)
print("Feature Importances:\n", feature_importances)
```

```python
# Graficamos para tener una visulizacion clara.

importances = rf_model.feature_importances_
features = X_train.columns

feature_importance_df = pd.DataFrame({'feature': features, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feature_importance_df, palette="pastel", hue='feature', dodge=False)
plt.title('Importancia de las Características')
plt.xlabel('Importancia')
plt.ylabel('Características')
plt.legend([],[], frameon=False)
plt.show()
```

Las características Online boarding, Inflight wifi service y Class son las más influyentes en la predicción de la satisfacción del cliente.

```python
# Matriz de Confusión (Confusion Matrix)
# Buscamos una forma visual de evaluar el rendimiento del modelo en términos de verdaderos positivos, falsos positivos, verdaderos negativos y falsos negativos.

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_pred = rf_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# Graficamos la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusión')
plt.show()
```

```python
# Evaluo el modelo y muestro el informe de clasificaciòn
y_pred = rf_model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
```

# Conclusiones generales

- Las características `Online boarding`, `Inflight wifi service` y `Class` son las más influyentes en la predicción de la satisfacción del cliente.
- El modelo de clasificación tiene un rendimiento muy sólido con una precision, recall y f1-score alrededor del 94%. Este rendimiento es consistente en ambas clases, lo que indica un buen equilibrio y una capacidad general de hacer predicciones precisas.
