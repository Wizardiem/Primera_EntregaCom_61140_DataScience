```python
# Sabemos que las primeras 2 columnas no aportan al dataset, procedemos a eliminarlas
df = df.drop(['Unnamed: 0', 'id'], axis=1)
# Comprobamos el resultado
df.info()
```

# Estrategias para manejar valores en 0 y vacios

```python
# Por descripcion de columnas sabemos cuales permiten valores en 0, lo cual significa que no fue evaluado por el pasajero
# Para estos casos definimos pasar los valores en 0 a nam para luego manejar los nam.
# Primero creamos una lista con todos los nombres de columna donde podemos realizar esta accion
satisfaction_cols = [
    'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking',
    'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment',
    'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service',
    'Inflight service', 'Cleanliness'
]
# Luego reemplazamos los valores 0 por NaN
df[satisfaction_cols] = df[satisfaction_cols].replace(0, np.nan)

#Luego de realizar el cambio, procedemos a verificar cuantos nam hay por columna para determinar la estrategia a seguir.
df.isna().sum()
```

```python
# Definimos como estrategias específicas para manejar NaN reemplazarlos por la media de todas las columnas ya que en ningun caso supera el 5% de la cantidad de datos.

#Guardamos la media de cada columna como INT
mean_values = df[satisfaction_cols].mean().round().astype(int)

#Reemplazmos los valores con las medias calculadas
for col in satisfaction_cols:
    df[col].fillna(mean_values[col], inplace=True)

# Verificamos el cambio
df.isna().sum()
```

```python
# Los unicos datos que nos quedan en NAM son de la columna Arrival Delay in minutes
# En este caso vamos a eliminar los registros donde la columna Arrival delay in minutes es nam

df.dropna(inplace=True)

# Verificamos el cambio
df.isna().sum()
```

## Analisis visual utilizando graficas

```python
# Procedemos a realizar graficas de torta para todos las columnas de Satisfaccion definidas anteriormente

plt.figure(figsize=(15, 20))
for i, col in enumerate(satisfaction_cols, 1):
    plt.subplot(5, 3, i)
    counts = df[col].value_counts()
    plt.pie(counts, labels=counts.index, colors=sns.color_palette("pastel"), autopct='%1.1f%%')
    plt.title(col)
plt.tight_layout()
plt.show()
```

Con el resultado de las graficas de torta podemos ver calaramente cuales son los atributos con mejor evaluaciones como por ejemplo Inflight service (nivel de satisfacción con el servicio a bordo) donde claramente el mayor porcentaje esta en las puntuaciones de 4 y 5.

Por otro lado Food and Dink (nivel de satisfacción con la comida y bebida)
Tiene un comportamiento mas parejo.




```python
# Siguiendo con el analisi de datos, realizamos un Histograma de las columnas Age y Flight Distance para entender donde se encuentran la mayor proporcion
# de edades a las que se le hizo la encuesta y cuales son las distancias que mas se repiten.

numeric_cols = ['Age', 'Flight Distance']


plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

Podemos determinar que la mayor cantidad de personas que viajan estan entre los 35 y 50 años y los viajes mas evaluados son aquellos de corta distancia.


```python

# Vamos a visualizar la relacion entre Genero y la Satisfaccion

contingency_table = pd.crosstab(df['satisfaction'], df['Gender'])
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap='coolwarm', fmt='d')
plt.title('Relación entre Satisfacción y Género')
plt.xlabel('Género')
plt.ylabel('Satisfacción')
plt.show()
```

Con el mapa de calor podemos determinar muy rapidamente que las mujeres son las que tienen un nivel de Satisfaccion en neutral o insastisfecha.

```python
#Evaluemos la satisfaccion por el clase del pasaje

plt.figure(figsize=(8, 4))
sns.displot(data=df, x='satisfaction', hue='Class', col='Class', palette='pastel', kde=True)
plt.suptitle('Distribución de la Satisfacción General por Clase de Viaje', y=1.05)
plt.show()

```

Vemos una clara evidencia de que los pasajes Business obtienen mejor calificacion, totalmente lo contrario con los pasajes en clase Eco. Con respecto a Eco plus ya la relacion es mas pareja, de todas maneras tiene mayor resultado la clasificacion neutra o no satisfecha.

```python
# Evaluemos la satisfacción general entre tipos de clientes

plt.figure(figsize=(8, 4))
sns.displot(data=df, x='satisfaction', hue='Customer Type', col='Customer Type', palette='pastel', kde=True)
plt.suptitle('Satisfacción General por Tipo de Cliente', y=1.05)
plt.show()
```

En los clientes no leales es donde el resultado Neutral o no satisfecho se acentua mas. En los clientes leales la realacion es mas pareja.
